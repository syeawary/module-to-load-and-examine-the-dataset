```python
# Using CSV module
# Project 1
# TEAM MEMBER NAMES
#Kenny Giang  CWID: 890062292
# Suhrab Yeawary CWIID: 888200631
import csv

temp_0 = []
temp_1 = []
temp_2 = []
temp_3 = []

week_1 = []
week_2 = []
week_3 = []

weeks = []
roll = []

result = []
#**************** please replace this with "participants.csv" when opening the .csv
with open("C:\\Users\sohra\Downloads\\participants.csv", newline='') as f:
    file = csv.reader(f)
    for line in file:
        temp_0.append(line)
        temp_1.append(line[1])
        temp_2.append(line[2])
        temp_3.append(line[3])

for line in temp_0[1::]:
    roll.append(line)
del temp_0[:]
for line in temp_1[1::]:
    week_1.append(line)
del temp_1[:]
for line in temp_2[1::]:
    week_2.append(line)
del temp_2[:]
for line in temp_3[1::]:
    week_3.append(line)
del temp_3[:]

weeks.append(week_1)
weeks.append(week_2)
weeks.append(week_3)

def tukey_range(q1, q3):
    return [q1 - (1.5 * (q3 - q1)), q3 + (1.5 * (q3 - q1))]

def tardy(students, week, roll):
    result = list(map(int, students))
    result.sort()
    index = 0
    
    i = len(result)
    j = int(i / 2)
    if i % 2 == 0:
        sum = result[j] + result[j - 1]
        q2 = sum / 2
        lower = j // 2
        upper = j + lower
        lower = (result[lower - 1] + result[lower]) / 2
        upper = (result[upper - 1] + result[upper]) / 2
    else:
        q2 = result[j]

    print("Week ", week,  "\n---------")
    print("Q1 is: ", lower)
    print("Q2 is: ", q2)
    print("Q3 is: ", upper)
    
    print("\nTukey Range:")
    outliers = tukey_range(lower, upper)
    low_bound = outliers[0]
    i = 0
    for item in outliers:
        if i == 0:
            print("Lower Bound: ", item)
        else:
            print("Upper Bound: ", item)
        i += 1
    del result[:]
    
    for time in students:
        check = int(time)
        if check < low_bound:
            print("Student: ", roll[index][0], "Time: ", roll[index][week])
        index += 1
        
    print("---------\n\n")

for x in range (3):
    tardy(weeks[x], x + 1, roll)
```
